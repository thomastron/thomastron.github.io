# Calculating Interest
**Simple Interest** is calculated just one time.
$$
\boxed{A = P (1 + r)^t}
$$
where:
- **A** - final amount
- **P** - principal (starting balance)
- **r** - annual interest rate (as a decimal)
- **t** - time (in years)

**Periodically Compounding Interest** is expressed as a fucntion of the number of times ($n$) the interst is applied:
- **n** - number of compounds per year

$$
\boxed{A = P \left(1 + \frac{r}{n}\right)^{nt}}
$$



>Bernoulli noticed that this sequence approaches a limit (the force of interest) with larger n and, thus, smaller compounding intervals. Compounding weekly ```(n = 52) yields $2.692596...```, while compounding daily ```(n = 365) yields $2.714567... ```
(approximately two cents more). The limit as n grows large is the number that came to be known as **e** Euler's Number. That is... with **Continuously Compounding Interest**, the account value will reach ```$2.718281828...```
$$
\begin{aligned}
1 (1 + \frac{1}{1})^{1*1} &= 2^1 = 2 \\
1 (1 + \frac{1}{2})^{2*1} &= 1.5^2 = 2.25 \\
1 (1 + \frac{1}{4})^{4*1} &= 1.25^4 = 2.4414  \\
1 (1 + \frac{1}{365})^{365*1} &= 1.002739^{365} = 2.714576
\end{aligned}
$$
$$
e = \lim_{n \to \infty} \sum_{i=1}^{n} (1 + \frac{r}{n})^{nt}
$$
while $r=1$ and $t=1$ and after some simplification and clever rearranging...
$$
\boxed{e = \sum_{n=0}^{\infty}\frac{1}{n!}} = 1 + \frac{1}{1!} + \frac{1}{2!} + \frac{1}{3!} + \ldots
$$








so the **Continuously Compounding Interest** formula is:

$$
\boxed{A=P e^{rt}}
$$
